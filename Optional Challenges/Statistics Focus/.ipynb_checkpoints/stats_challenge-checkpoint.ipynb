```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
orders = pd.read_csv('orders.csv')

plt.boxplot(orders['order_value'])
plt.title('Distribution of Order Values')
plt.ylabel('Order Value')
plt.show()
```

I mean just looking at this, its pretty obvious that there is either one or several outliers which are $700,000 orders, this is throwing the average of the data off completely since most of the order are much much less. You can see the average is close to zero compared to the enourmous upper limit of order values. 

Intuitively a better measure for a data set like this would be to use a median so that we can take into account the an order value that is at contained within the majority of the data, (50th percentile).

```python
orders['order_value'].median()
```

As we can see, the median is much much more reasonable of what we would expect from the average order price of sneakers, 284 is more realistic for out context. 

To go further I want to explore different percentiles of the data.

```python
print("1st - 4th percentile\n")
print(orders['order_value'].quantile(0.001))
print(orders['order_value'].quantile(0.01))
print(orders['order_value'].quantile(0.02))
print(orders['order_value'].quantile(0.03))
print(orders['order_value'].quantile(0.04))
print("96th - 99.9th percentile\n")
print(orders['order_value'].quantile(0.96))
print(orders['order_value'].quantile(0.97))
print(orders['order_value'].quantile(0.98))
print(orders['order_value'].quantile(0.99))
print(orders['order_value'].quantile(0.999))
```

From this you can really see how the numbers increase dramatically towards the upper 5 percentiles, but the same is not seen in the bottom 5. This means some unusual activity is happening towards upper bound. Maybe people are actually buying $200,000 - $700,000 shoes occasionally. 

```python
# Convert the 'created_at' column to a datetime format
orders['created_at'] = pd.to_datetime(orders['created_at'])

# pandas magic mdst taught me
orders['day'] = orders['created_at'].dt.date
orders['hour'] = orders['created_at'].dt.hour
daily_orders = orders.groupby('day').size()
hourly_orders = orders.groupby('hour').size()

#  daily orders

daily_orders.plot()
plt.title('Number of Orders per Day')
plt.ylabel('Orders')
plt.xlabel('Date')
plt.show()

# hourly orders
hourly_orders.plot(kind='line')
plt.title('Number of Orders for Hour')
plt.ylabel('Number of Orders')
plt.xlabel('Hour')
plt.show()
```

This doesn't really reveal much, I think these visualizations of when people buy shoes is pretty normal, theres no specific time or day that unusual activity is occurring. 

```python
Q1 = orders['order_value'].quantile(0.25)
Q3 = orders['order_value'].quantile(0.75)

IQR = Q3 - Q1
print(IQR)
```

```python
# finding all orders in the middle 50% of the data
filtered_orders = orders[(orders['order_value'] >= Q1) & (orders['order_value'] <= Q3)]
# averaging from this data
print("average:" , filtered_orders['order_value'].mean())
```

This reinforces our observations from before that a more resonable figure for a typical order value is $250 - $300, and definetly not $3145, since this figure is offset by outliers.
